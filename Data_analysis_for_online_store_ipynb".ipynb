```python
# Connecting Google Drive
from google.colab import drive
drive.mount("/content/drive")

%cd /content/drive/MyDrive/Colab\ Notebooks

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_countries = pd.read_csv("countries.csv")
df_events = pd.read_csv("events.csv")
df_products = pd.read_csv("products.csv")

df_countries = df_countries.dropna()
df_events = df_events.dropna()
df_products = df_products.dropna()

df_events["Order Date"] = pd.to_datetime(df_events["Order Date"])
df_events["Ship Date"] = pd.to_datetime(df_events["Ship Date"])

country_event_merged = pd.merge(df_countries, df_events, left_on="alpha-3", right_on="Country Code", how="inner")
df = pd.merge(country_event_merged , df_products , left_on="Product ID" , right_on="id" , how = "inner")

df = df.drop("alpha-2" , axis = 1)
df = df.drop("alpha-3" , axis = 1)
df = df.drop("id" , axis = 1)
df.rename(columns={"name": "country"}, inplace=True)

df["Sales Channel"] = df["Sales Channel"].replace("online", "Online")

df["Revenue"] = df["Units Sold"] * df["Unit Price"]
df["Cost"] = df["Units Sold"] * df["Unit Cost"]
df["Profit"] = df["Revenue"] - df["Cost"]

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].barh(df["item_type"], df["Profit"], color="blue")
ax[0, 0].set_title("Profit by Category")
ax[0, 0].set_xlabel("Category")
ax[0, 0].set_ylabel("Profit")

ax[0, 1].barh(df["sub-region"], df["Profit"], color="blue")
ax[0, 1].set_title("Profit by Sub-Region")
ax[0, 1].set_xlabel("Sub-Region")
ax[0, 1].set_ylabel("Profit")

ax[1, 0].bar(df["Sales Channel"], df["Profit"], color="blue")
ax[1, 0].set_title("Profit by Sales Channel")
ax[1, 0].set_xlabel("Sales Channel")
ax[1, 0].set_ylabel("Profit")

ax[1, 1].bar(df["region"], df["Profit"], color="blue")
ax[1, 1].set_title("Profit by Region")
ax[1, 1].set_xlabel("Reion")
ax[1, 1].set_ylabel("Profit")

plt.tight_layout()

plt.show()
```

##### З графіку Profit by Category видно, що найбільше прибутку приносять товари категорій Cosmetics та Household, а найменше Fruits та Beverages.
##### Згідно графіку Profit by Sub-Region, прибуток в суб-регіонах Європи приблизно однаковий, а Західна Азія приносить найменше прибутку.
##### На графіку Profit by Sales Channel видно, що прибуток майже рівний між продажами Offline та Online
##### А з графку Profit by Region втдно, що прибуток від продажів у Європі відчутно перевищує прибуток від продажів у Азії.

```python
# Connecting Google Drive
from google.colab import drive
drive.mount("/content/drive")

%cd /content/drive/MyDrive/Colab\ Notebooks

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_countries = pd.read_csv("countries.csv")
df_events = pd.read_csv("events.csv")
df_products = pd.read_csv("products.csv")

df_countries = df_countries.dropna()
df_events = df_events.dropna()
df_products = df_products.dropna()

df_events["Order Date"] = pd.to_datetime(df_events["Order Date"])
df_events["Ship Date"] = pd.to_datetime(df_events["Ship Date"])

country_event_merged = pd.merge(df_countries, df_events, left_on="alpha-3", right_on="Country Code", how="inner")
df = pd.merge(country_event_merged , df_products , left_on="Product ID" , right_on="id" , how = "inner")

df = df.drop("alpha-2" , axis = 1)
df = df.drop("alpha-3" , axis = 1)
df = df.drop("id" , axis = 1)
df.rename(columns={"name": "country"}, inplace=True)

df["Sales Channel"] = df["Sales Channel"].replace("online", "Online")

df["Revenue"] = df["Units Sold"] * df["Unit Price"]
df["Cost"] = df["Units Sold"] * df["Unit Cost"]
df["Profit"] = df["Revenue"] - df["Cost"]
df["Order_ship_gap"] = (df["Ship Date"] - df["Order Date"]).dt.days

#df.head(10)

fig, ax = plt.subplots(2, 2, figsize=(12, 12))

ax[0, 0].barh(df["country"], df["Order_ship_gap"], color="blue")
ax[0, 0].set_title("Days before shipment by country")
ax[0, 0].set_xlabel("Country")
ax[0, 0].set_ylabel("Days before shipment")

ax[0, 1].barh(df["sub-region"], df["Order_ship_gap"], color="blue")
ax[0, 1].set_title("Days before shipment by Sub-Region")
ax[0, 1].set_xlabel("Sub-Region")
ax[0, 1].set_ylabel("Days before shipment")

ax[1, 0].barh(df["item_type"], df["Order_ship_gap"], color="blue")
ax[1, 0].set_title("Days before shipment by Category")
ax[1, 0].set_xlabel("Category")
ax[1, 0].set_ylabel("Days before shipment")

ax[1, 1].bar(df["region"], df["Order_ship_gap"], color="blue")
ax[1, 1].set_title("Days before shipment by Region")
ax[1, 1].set_xlabel("Region")
ax[1, 1].set_ylabel("Days before shipment")

plt.tight_layout()

plt.show()
```

```python
# Connecting Google Drive
from google.colab import drive
drive.mount("/content/drive")

%cd /content/drive/MyDrive/Colab\ Notebooks

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_countries = pd.read_csv("countries.csv")
df_events = pd.read_csv("events.csv")
df_products = pd.read_csv("products.csv")

df_countries = df_countries.dropna()
df_events = df_events.dropna()
df_products = df_products.dropna()

df_events["Order Date"] = pd.to_datetime(df_events["Order Date"])
df_events["Ship Date"] = pd.to_datetime(df_events["Ship Date"])

country_event_merged = pd.merge(df_countries, df_events, left_on="alpha-3", right_on="Country Code", how="inner")
df = pd.merge(country_event_merged , df_products , left_on="Product ID" , right_on="id" , how = "inner")

df = df.drop("alpha-2" , axis = 1)
df = df.drop("alpha-3" , axis = 1)
df = df.drop("id" , axis = 1)
df.rename(columns={"name": "country"}, inplace=True)

df["Sales Channel"] = df["Sales Channel"].replace("online", "Online")

df["Revenue"] = df["Units Sold"] * df["Unit Price"]
df["Cost"] = df["Units Sold"] * df["Unit Cost"]
df["Profit"] = df["Revenue"] - df["Cost"]
df["Order_ship_gap"] = (df["Ship Date"] - df["Order Date"]).dt.days
df["Order_weekday"] = df["Order Date"].dt.day_name()

group_weekday = df.groupby(["item_type" , "Order_weekday"])["Profit"].sum().reset_index()
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
group_weekday["Order_weekday"] = pd.Categorical(group_weekday["Order_weekday"], categories=weekday_order, ordered=True)

sns.catplot(x="Order_weekday", y="Profit", kind="bar" , data=group_weekday, hue="item_type" , height=5, aspect=2)
plt.title("Profit by Categories")
plt.xlabel("Weekday")
plt.ylabel("Profit")
plt.show()
```
